# Load Data

```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. LOAD DATA
nama_file = 'W9371UM ULTRASONIK_6-9 November 2025.xlsx'

# Melewati 3 baris teratas (sampah) dan memuat tanpa header
data = pd.read_excel(nama_file,)

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# 2. DATA CLEANING - Convert VALUE_FUEL to integer
col = 'VALUE_FUEL'
if col in data.columns:
    cleaned = data[col].astype(str).str.replace(r"[,\s$â‚¬Â£]", "", regex=True)
    data[col] = pd.to_numeric(cleaned, errors='coerce')
    data[col] = data[col].round().astype('Int64')
else:
    print(f"Column '{col}' not found")

# 3. FUEL VALUE ANALYSIS
print("FUEL VALUE ANALYSIS")

# Basic statistics
print("\n1. BASIC STATISTICS")
print(data['VALUE_FUEL'].describe())

# Missing values
print(f"\n2. MISSING VALUES: {data['VALUE_FUEL'].isna().sum()} ({data['VALUE_FUEL'].isna().sum()/len(data)*100:.2f}%)")

# Distribution analysis
print("\n3. DISTRIBUTION")
print(f"Min: {data['VALUE_FUEL'].min()}")
print(f"Max: {data['VALUE_FUEL'].max()}")
print(f"Mean: {data['VALUE_FUEL'].mean():.2f}")
print(f"Median: {data['VALUE_FUEL'].median():.2f}")
print(f"Std Dev: {data['VALUE_FUEL'].std():.2f}")

# Value counts (top 10)
print("\n4. TOP 10 FUEL VALUES")
print(data['VALUE_FUEL'].value_counts().head(10))

# Visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Histogram
axes[0, 0].hist(data['VALUE_FUEL'].dropna(), bins=30, edgecolor='black', color='skyblue')
axes[0, 0].set_title('Distribution of Fuel Values')
axes[0, 0].set_xlabel('Value Fuel')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(alpha=0.3)

# Box plot
axes[0, 1].boxplot(data['VALUE_FUEL'].dropna())
axes[0, 1].set_title('Box Plot of Fuel Values')
axes[0, 1].set_ylabel('Value Fuel')
axes[0, 1].grid(alpha=0.3)

# Time series (if datetime column exists)
if 'Tanggal' in data.columns or 'DATE' in data.columns:
    date_col = [col for col in data.columns if 'Tanggal' in col or 'DATE' in col][0]
    axes[1, 0].plot(data[date_col], data['VALUE_FUEL'], marker='o', linestyle='-', linewidth=1)
    axes[1, 0].set_title('Fuel Value Over Time')
    axes[1, 0].set_xlabel('Date')
    axes[1, 0].set_ylabel('Value Fuel')
    axes[1, 0].tick_params(axis='x', rotation=45)
    axes[1, 0].grid(alpha=0.3)

# Quantile analysis
quantiles = data['VALUE_FUEL'].quantile([0.25, 0.5, 0.75, 0.9, 0.95, 0.99])
axes[1, 1].bar(quantiles.index.astype(str), quantiles.values, color='coral', edgecolor='black')
axes[1, 1].set_title('Quantile Analysis')
axes[1, 1].set_xlabel('Quantile')
axes[1, 1].set_ylabel('Value Fuel')
axes[1, 1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
```

```python
# Group by city and sum fuel values
fuel_by_city = data.groupby('LOCATION')['VALUE_FUEL'].sum().sort_values(ascending=False)

print("FUEL USAGE BY LOCATION")
print(fuel_by_city)

# Visualization
plt.figure(figsize=(12, 6))
fuel_by_city.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Total Fuel Usage by City')
plt.xlabel('City')
plt.ylabel('Total Fuel Value')
plt.xticks(rotation=45, ha='right')
plt.grid(alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```
